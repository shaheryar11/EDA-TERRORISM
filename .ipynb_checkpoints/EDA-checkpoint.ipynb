```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline 
```

IMPORTING DATA 


```python
data=pd.read_csv('globalterrorismdb_0718dist.csv' , encoding='latin1')
df=pd.DataFrame(data)
print("data loaded")
```

# CLEANING THE DATA

```python
df=df[["iyear","imonth","iday","country_txt","region_txt","provstate","city",
      "latitude","longitude","location","summary","attacktype1_txt","targtype1_txt","gname","motive","weaptype1_txt","nkill",
      "nwound","addnotes"]]
```

```python
df.head()
```

```python
df.rename(columns={"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country","region_txt":"Region",
                  "provstate":"Province/State","city":"City","latitude":"Latitude","longitude":"Longitude","location":"Location",
                  "summary":"Summary","attacktype1_txt":"Attack Type","targtype1_txt":"Target Type","gname":"Group Name","motive":"Motive",
                  "weaptype1_txt":"Weapon Type","nkill":"Killed","nwound":"Wound","addnotes":"NOTES"},inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Killed"]=df["Killed"].fillna(0)
df["Wound"]=df["Wound"].fillna(0)
df["Casualty"]=df["Killed"]+df["Wound"]
```

```python
df.describe()
```

# OBSERVATION
1.DATA FROM YEAR 1970 TO 2017
2.MAX PEOPLE KILLED : 1570,
3.MAX PEOPLE WOUNDED : 8191,
4.MAX CASUALTIES : 9574,

# 1.ATTACKS EACH YEAR

```python
attacks=df["Year"].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={"index":"Year","Year":"Attacks"}).set_index("Year")
```

```python
attacks.head(10)
```

```python
attacks.plot(kind="bar",color="red",figsize=(15,6),fontsize=13)
plt.title("ATTACKS EACH YEAR")
plt.xlabel("Years",fontsize=14)
plt.ylabel("No. of Attacks ",fontsize=14)
plt.show()
```

Total Casualties in each year

```python
cas_yr=df[["Year","Casualty"]].groupby("Year").sum()
cas_yr.head()
```

```python
cas_yr.plot(kind="bar",color="blue",figsize=(15,6))
plt.xlabel("Years",fontsize=13)
plt.ylabel("NUMBER OF CASUALTIES")
plt.show()
```

```python
kill_yr=df[["Year","Killed"]].groupby("Year").sum()
kill_yr.head()
```

```python
kill_yr.plot(kind="bar",color="grey",figsize=(15,6))
plt.title("PEOPLE KILLED IN EACH YEAR")
plt.xlabel("Years",fontsize=13)
plt.ylabel("NUMBER OF PEOPLE KILLED")
plt.show()
```

```python
wou_yr=df[["Year","Wound"]].groupby("Year").sum()
wou_yr.head()
```

```python
wou_yr.plot(kind="bar",color="yellow",figsize=(15,6))
plt.title("PEOPLE WOUNDED IN EACH YEAR")
plt.xlabel("Years",fontsize=13)
plt.ylabel("NUMBER OF PEOPLE WOUNDED")
plt.show()
```

# 2.REGION WISE ATTACK

1.Terrorist Attack (Region Wise) from Year 1970 to 2017

```python
rg=pd.crosstab(df.Year,df.Region) #crosstab for frequency table
rg.head()
```

```python
rg.plot(kind="area", stacked=False ,alpha=0.5,figsize=(20,8))
plt.title("REGION WISE ATTACKS",fontsize=15)
plt.xlabel("Years",fontsize=15)
plt.ylabel("NO OF ATTACKS",fontsize=15)
plt.show()
```

2.TOTAL ATTACKS IN EACH REGION

```python
regt=rg.transpose()
regt["Total"]=regt.sum(axis=1)
ra=regt["Total"].sort_values(ascending=False)
ra
```

```python
ra.plot(kind="bar",figsize=(15,6))
plt.title("Total Number of Attacks in each Region from 1970-2017")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```

3.Total Casualties in Each Region


```python
rc=df[["Region","Casualty"]].groupby("Region").sum().sort_values(by="Casualty",ascending=False)
rc
```

```python
rc.plot(kind="bar",color="cornflowerblue",figsize=(15,6))
plt.title("Region wise Casualties",fontsize=13)
plt.xlabel("Regions",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Casualties",fontsize=13)
plt.show()
```

4.Killed In each Region

```python
rk=df[["Region","Killed"]].groupby("Region").sum().sort_values(by="Killed",ascending=False)
rk
```

5.Wounded in Each Region

```python
rw=df[["Region","Wound"]].groupby("Region").sum().sort_values(by="Wound",ascending=False)
rw
```

```python
fig=plt.figure()
ax0=fig.add_subplot(1,2,1)
ax1=fig.add_subplot(1,2,2)

#Killed
rk.plot(kind="bar",color="cornflowerblue",figsize=(15,6),ax=ax0)
ax0.set_title("People Killed in each Region")
ax0.set_xlabel("Regions")
ax0.set_ylabel("Number of People Killed")

#Wounded
rw.plot(kind="bar",color="cornflowerblue",figsize=(15,6),ax=ax1)
ax1.set_title("People Wounded in each Region")
ax1.set_xlabel("Regions")
ax1.set_ylabel("Number of People Wounded")

plt.show()
```
